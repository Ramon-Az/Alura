```python
# 13) As pessoas colaboradoras de um setor da empresa que você trabalha vão receber um abono correspondente a 10% do salário devido ao ótimo desempenho do time. O setor financeiro solicitou sua ajuda para a verificação das consequências financeiras que esse abono irá gerar nos recursos. Assim, foi encaminhada para você uma lista com os salários que receberão o abono: [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]. O abono de cada colaborador(a) não pode ser inferior a 200. Em código, transforme cada um dos salários em chaves de um dicionário e o abono de cada salário no elemento. Depois, informe o total de gastos com o abono, quantos(as) colaboradores(as) receberam o abono mínimo e qual o maior valor de abono fornecido.

salarios = [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]
abonos = {}
for salario in salarios:
    abono = max(200, salario * 0.1)
    abonos[salario] = abono
    
total_abonos = sum(abonos.values())
colaboradores_abono_minimo = sum(1 for abono in abonos.values() if abono == 200)
maior_abono = max(abonos.values())

print(f"Total de gastos com o abono: R$ {total_abonos:.2f}")
print(f"Quantidade de colaboradores que receberam o abono mínimo: {colaboradores_abono_minimo}")
print(f"Maior valor de abono fornecido: R$ {maior_abono:.2f}")
```

```python
# 14) Uma equipe de cientistas de dados está estudando a diversidade biológica em uma floresta. A equipe fez a coleta de informações sobre o número de espécies de plantas e animais em cada área dessa floresta e armazenou essas informações em um dicionário. Nele, a chave descreve a área dos dados e os valores nas listas correspondem às espécies de plantas e animais nas áreas, respectivamente.

dados_coletados = {
    "Área Norte": [2819, 7236],
    "Área Leste": [1440, 9492],
    "Área Sul": [5969, 7496],
    "Área Oeste": [14446, 49688],
    "Área Centro": [22558, 451448]
}

total_plantas = sum(dados[0] for dados in dados_coletados.values())
total_animais = sum(dados[1] for dados in dados_coletados.values())

media_especie_norte = sum(dados_coletados["Área Norte"]) / 2
media_especie_leste = sum(dados_coletados["Área Leste"]) / 2
media_especie_sul = sum(dados_coletados["Área Sul"]) / 2
media_especie_oeste = sum(dados_coletados["Área Oeste"]) / 2
media_especie_centro = sum(dados_coletados["Área Centro"]) / 2

area_mais_diversa = max(dados_coletados, key=lambda area: sum(dados_coletados[area]))

print(f"Total de espécies de plantas: {total_plantas}")
print(f"Total de espécies de animais: {total_animais}")

print(f"Média de espécies na Área Norte: {media_especie_norte:.2f}")
print(f"Média de espécies na Área Leste: {media_especie_leste:.2f}")
print(f"Média de espécies na Área Sul: {media_especie_sul:.2f}")
print(f"Média de espécies na Área Oeste: {media_especie_oeste:.2f}")
print(f"Média de espécies na Área Centro: {media_especie_centro:.2f}")

print(f"Área mais diversa: {area_mais_diversa}")
```

```python
# 15) O setor de RH da sua empresa te pediu uma ajuda para analisar as idades de colaboradores(as) de 4 setores da empresa. Para isso, foram fornecidos os seguintes dados:

idades_colaboradores = {
    "setor A": [22, 26, 30, 35, 38, 40, 56, 57, 65],
    "setor B": [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
    "setor C": [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
    "setor D": [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
}
# Sabendo que cada setor tem 10 colaboradores(as), construa um código que calcule a média de idade de cada setor, a idade média geral entre todos os setores e quantas pessoas estão acima da idade média geral.

media_idades_setor = {setor: sum(idades) / len(idades) for setor, idades in idades_colaboradores.items()}

idades_geral = [idade for idades in idades_colaboradores.values() for idade in idades]

media_geral = sum(idades_geral) / len(idades_geral)

pessoas_acima_media = sum(1 for idade in idades_geral if idade > media_geral)

print("Médias de idade por setor:")
for setor, media in media_idades_setor.items():
    print(f"{setor}: {media:.2f}")

print(f"\nIdade média geral: {media_geral:.2f}")
print(f"Número de pessoas acima da média geral: {pessoas_acima_media}")
```
